DATA CLEANING
Finding null values in the dataset

```python
 # splitting a series in two and get the nemericall part
car_data['Engine'] = car_data['Engine'].str.split().str.get(0)

car_data['Max Power'] = car_data['Max Power'].str.split().str.get(0)
#replacing characters in a series
car_data['Max Power'] = car_data['Max Power'].str.replace("@"," ")
 #getting the first index as Hp
car_data['Max Power'] = car_data['Max Power'].str.split().str.get(0)
#converting object to type float
car_data ['Max Power'] = car_data ['Max Power'].astype(float).round()

```

```python
car_data.isnull()
```

```python
car_data.isnull().sum()
```

```python
car_data['Engine'].fillna(car_data['Engine'].mode(),inplace = True)
```

```python
car_data['Seating Capacity'].fillna(car_data['Seating Capacity'].mean(),inplace = True)
```

```python
car_data['Height'].fillna(car_data['Height'].mean(),inplace = True)
```

```python
#car_data['Width'].fillna(car_data['Width'].mean(),inplace = True)
car_data.dropna(subset=["Width"], axis=0, inplace=True)
```

```python
#car_data['Length'].fillna(car_data['Length'].mean(),inplace = True)
# droping null
car_data.dropna(subset=["Length"], axis=0, inplace=True)
```

```python
car_data['Fuel Tank Capacity'].fillna(car_data['Fuel Tank Capacity'].mode(),inplace = True)
```

```python
car_data.dropna(subset=["Drivetrain"], axis=0, inplace=True)
```

```python
car_data.dropna(subset=["Engine"], axis=0, inplace=True)
```

```python
car_data.dropna(subset=["Fuel Tank Capacity"], axis=0, inplace=True)
```

create line plot for cars from Mumbai','Delhi','Bangalore','Pune','Hyderabad' relative to their prices

```python
#create a figure with larger 
plt.figure(figsize= (12,8))

# calculate average price for cars from 'Mumbai','Delhi','Bangalore','Hyderabad']
avg_price= num_byloc = car_data[car_data['Location']. isin(['Mumbai','Delhi','Pune','Bangalore','Hyderabad'])] ['Price']. mean()
print(avg_price)

data = car_data[car_data['Location']. isin(['Mumbai','Delhi','Pune','Bangalore','Hyderabad'])]
sns.lineplot(x='Make', y='Price', data = data , color='skyblue', marker='o', hue= 'Location')

# add a horizontal line to indicate the average price
plt.axhline(y=avg_price, color='black', linestyle='dashed')

#set labels and title
plt.title('Average Car Prices by Make (Mumbai ,Delhi,Pune, Bangalore,and Hyderabad)', fontsize=13)
plt.xlabel('Make',fontsize=11)
plt.ylabel('Price',fontsize=11)
plt.legend(loc='upper left',fontsize=11)
plt.xticks(rotation=45) # to rotate x_axis label by 45 degrees
plt.show()
```

```python
# Sort the dataframe by Price in descending order
car_data = car_data.sort_values(by='Price',ascending=False)

# Select the top 15 car types
car_expensive= car_data.head(15)

# Create a bar plot
sns.barplot(x = 'Make', y = 'Price', data = car_expensive, hue = 'Location',palette='Blues')


# Set labels and titles
plt.title("Price Distribution of Top 6 Car Types by Make and Origin", fontsize=13)
plt.xlabel("Make",fontsize=10)
plt.ylabel("Price",fontsize=10)
plt.xticks(rotation=90) # to rotate x_axis label by 90 degrees)
plt.show()
```

```python
sns.set_style("whitegrid")
# create a bar plot with larger size
plt.figure(figsize=(13,7))
sns.barplot(x = car_expensive['Make'], y = 'Price', data = car_expensive, hue = 'Location', palette='Blues')

#set labels and titles
plt.title("Price Distribution by No. of Make", fontsize=14)
plt.xlabel("Make",fontsize=8)
plt.ylabel("Price",fontsize=8)
plt.xticks(rotation=45) # to rotate x_axis label by 45 degrees
plt.show()
```

There is a positive relationship between weight and horsepower, meaning that as the weight of a car increases, so does its horsepower.
Some European car manufacturers have models that defy the trend of the relationship between weight and horsepower, as they have cars that weigh less than 4500 pounds and have a horsepower of nearly 500.

```python
#group the data by car type and calculate the average fuel tank capacity
car_data_grouped = car_data.groupby('Make')[['Fuel Tank Capacity']] .mean()

#sort the data by average fuel tank capacity size in descending order
car_data_grouped.sort_values(by="Fuel Tank Capacity" , ascending= False, inplace = True)

#select the top 10 car types by average fuel tank capaciity
top_10_cars = car_data_grouped.head(10)

#plot size
plt.figure(figsize=(10,7))
#create a bar plot of the average cylinder size by top 10 car types
sns.barplot(x= top_10_cars . index , y= 'Fuel Tank Capacity' ,  palette='PuBu', data = top_10_cars )

#set the title and labels
plt.title( "average fuel size by top 10 car types" , fontsize = 14)
plt.xlabel("carn type" ,fontsize = 14)
plt.ylabel("avg fuel tank capacity" ,fontsize = 14)
plt.xticks(rotation = 45) # to rotate x_axis label by 45 degrees
plt.show()
```
